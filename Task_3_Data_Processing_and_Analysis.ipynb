```python
# Problem 1: Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# highest performance score
highest_score = max(employees.values())

# Identify all employees with the highest score
top_performers = [name for name, score in employees.items() if score == highest_score]


print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

```python
# Problem 2: Search Query Keyword Analysis

import string
query = "Buy mobile phone buy phone online"

# Converting to lowercase
query = query.lower()

# Removing punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Splitting into words
words = query.split()

# Counting word frequency
word_count = {}

for word in words:
    word_count[word] = word_count.get(word, 0) + 1

# Display only words searched more than once
result = {word: count for word, count in word_count.items() if count > 1}

print(result)
```

```python
# Problem 3: Sensor Data Validation

# Given sensor readings
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

# Identifying even readings and store (hour_index, value)
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((index, value))


print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Extracting domain and count occurrences
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Calculating total users
total_users = len(emails)

# Calculate and display percentage
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculating average sales
average_sales = sum(sales) / len(sales)

# Detecting sales more than 30% above average
threshold = average_sales * 1.3

# Display of spike days
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
# Problem 6: Duplicate User ID Detection

# Given user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

# Count occurrences of each ID
for user_id in user_ids:
    id_count[user_id] = id_count.get(user_id, 0) + 1

# Display duplicates only
for user_id, count in id_count.items():
    if count > 1:
        print(f"{user_id} → {count} times")
```
