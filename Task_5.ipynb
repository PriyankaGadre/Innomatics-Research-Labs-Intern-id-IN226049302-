```python
# Problem 1: Social Media – Post Engagement Analyzer

def analyze_post_engagement(likes_list):

    total_likes = 0

    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes

    # Conditional check for viral status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status

likes_data = [200, 300, 150, 400]
total, post_status = analyze_post_engagement(likes_data)

print("Total Likes:", total)
print("Post Status:", post_status)
```

```python
# Problem 2: Healthcare – Medicine Stock Alert System

def check_medicine_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    return stock, status


stock_value, stock_status = check_medicine_stock(6)

print("Medicine Stock:", stock_value)
print("Status:", stock_status)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker

def rainfall_checker(rainfall_data, required_level):

    total_rainfall = 0

    for rain in rainfall_data:
        total_rainfall += rain

    average_rainfall = total_rainfall / len(rainfall_data)

    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average_rainfall, status


rain_data = [60, 80, 70, 78]
avg_rain, rain_status = rainfall_checker(rain_data, 70)

print("Average Rainfall:", avg_rain)
print("Rainfall Status:", rain_status)
```

```python
# Problem 4: Social Media – Duplicate Account Detection

def detect_duplicate_accounts(usernames):

    unique_users = set(usernames)

    if len(unique_users) != len(usernames):
        result = "Yes"
    else:
        result = "No"

    return result

users = ["john123", "emma22", "john123", "alex99"]
duplicate_status = detect_duplicate_accounts(users)

print("Duplicate Accounts Found:", duplicate_status)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter

def filter_premium_crops(prices):

    premium_crops = []

    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    return premium_crops

crop_prices = [1500, 2500, 1800, 3200]
premium = filter_premium_crops(crop_prices)

print("Premium Crops:", premium)
```

```python
# Problem 7: System Monitoring – Application Health Checker

def application_health_checker(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    return error_count, status

# Eexample
errors, system_status = application_health_checker(7)

print("Error Count:", errors)
print("System Status:", system_status)
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker

def transaction_limit_checker(amount):
    daily_limit = 50000

    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    return amount, status

# Example
transaction_amount, transaction_status = transaction_limit_checker(60000)

print("Transaction Amount:", transaction_amount)
print("Transaction Status:", transaction_status)
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System

def attendance_eligibility(attendance_list):
    total_classes = len(attendance_list)
    attended_classes = 0

    for status in attendance_list:
        if status == 1:
            attended_classes += 1

    percentage = (attended_classes / total_classes) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    return percentage, eligibility


# example  ( 1 = Present, 0 = Absent )
attendance_data = [1, 1, 0, 1, 1]
attendance_percentage, exam_status = attendance_eligibility(attendance_data)

print("Attendance Percentage:", attendance_percentage)
print("Exam Eligibility:", exam_status)
```

```python
# Problem 10: Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
    bill = 0

    # calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    return bill, usage_status

total_bill, usage = calculate_electricity_bill(250)

print("Total Bill:", total_bill)
print("Usage Classification:", usage)
```
