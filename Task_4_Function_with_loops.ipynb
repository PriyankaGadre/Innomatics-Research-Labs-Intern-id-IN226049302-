```python
# Problem 1: Mobile Recharge Validation System

def mobile_recharge():

    available_plans = [199, 299, 399, 599]

    while True:
        try:
            amount = int(input("Enter recharge amount: ₹"))
            if amount < 50:
                print("Recharge must be at least ₹50. Try again.\n")
                continue
            if amount not in available_plans:
                print("Selected plan is not available. Try again.\n")
                continue
            print("Recharge completed successfully!")
            break
        except ValueError:
            print("Invalid input. Please enter numbers only.\n")

mobile_recharge()
```

```python
# Problem 2: Inventory Reorder Alert System

def inventory_status(products):

    for item in products:
        quantity = products[item]
        if quantity < 15:
            print(f"{item} -> Reorder Alert (Stock: {quantity})")
        else:
            print(f"{item} -> Stock OK (Stock: {quantity})")


# Sample dictionary
store_items = {
    "Printer": 12,
    "Scanner": 18,
    "Tablet": 9,
    "Speaker": 20
}

inventory_status(store_items)
```

```python
# Problem 3: Student Result Processing System

def student_result(marks):

    total_marks = 0

    # Adding marks using loop
    for score in marks:
        total_marks += score

    average_marks = total_marks / len(marks)

    print("Average Marks:", average_marks)

    if average_marks >= 50:
        return "Pass"
    else:
        return "Fail"

marks_list = [55, 60, 45, 70]
print("Final Result:", student_result(marks_list))
```

```python
# Problem 4: Cab Fare Estimator

def cab_fare(distance_km, is_peak):

    total = 50 + (12 * distance_km)
    if is_peak:
        total = total + (0.25 * total)

    return total

# Retry option
while True:
    try:
        distance = float(input("Enter travel distance (km): "))
        peak_time = input("Peak hour? (yes/no): ").strip().lower()

        peak_flag = True if peak_time == "yes" else False

        fare_amount = cab_fare(distance, peak_flag)
        print("Total Fare: ₹", fare_amount)
        break

    except ValueError:
        print("Please enter valid input.\n")
```

```python
# Problem 5: Attendance Eligibility Checker

def attendance_check(records):
    present_count = 0

    for status in records:
        if status == "P":
            present_count += 1

    percentage = (present_count / len(records)) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example data
attendance_data = ["P", "P", "A", "P", "P", "A", "P", "P"]
print("Status:", attendance_check(attendance_data))
```

```python
# Problem 6: Password Strength Checker

def password_validator(password):

    if len(password) < 8:
        return "Password is too short"

    digit_found = False
    special_found = False

    for ch in password:
        if ch.isdigit():
            digit_found = True
        if ch in "@#$":
            special_found = True

    if digit_found and special_found:
        return "Strong Password"
    else:
        return "Weak Password"


# Input from user
user_password = input("Enter password: ")
print(password_validator(user_password))
```
