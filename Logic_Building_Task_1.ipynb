```python
print("SECTION 2: Student Performance Analysis")
print("\n")

# Student marks dataset
student_marks = [45, 78, 90, 33, 60]

# Counter variables
students_passed = 0
students_failed = 0

# Analyzing each student's performance
print("Analyzing marks for 5 students...\n")

for current_marks in student_marks:
    if(current_marks >= 50):
        students_passed = students_passed + 1
        print(f" Student scored {current_marks} → PASS")
    else:
        students_failed = students_failed + 1
        print(f" Student scored {current_marks} → FAIL")

print(f" Total Students Passed: {students_passed}")
print(f" Total Students Failed: {students_failed}")
print(f" Class Strength: {len(student_marks)} students")


```

```python

print("SECTION 3: Data Cleaning & Standardization")
print("\n")

#user names
raw_names = [" Alice ", "bob", " CHARLIE "]

print("Before Cleaning:")
print(f" Raw Data: {raw_names}")
print(f" Issues: Extra spaces, inconsistent casing")

# Cleaning process
names = []

for raw_name in raw_names:
    # Removing leading spaces
    without_spaces = raw_name.strip()
    # Converting to lowercase
    lowercase_name = without_spaces.lower()
    # Adding to cleaned list
    names.append(lowercase_name)

print("After Cleaning:")
print(f" Cleaned Data: {names}")
print(f" Format: All lowercase, no extra spaces")
```

```python

print("SECTION 4: Message Length Analysis")
print("\n")

# Sample messages
chat_messages = ["Hi", "Welcome to the platform", "OK"]
limit = 10

print("Processing messages...")
print("\n")

message_counter = 1
for current_message in chat_messages:
    message_length = len(current_message)

    print(f"Message {message_counter}:")
    print(f"Content: \"{current_message}\"")
    print(f"Length: {message_length} characters")

    # Flagging long messages
    if(message_length > limit):
        print(f"Status: EXCEEDS LIMIT (>{limit} chars)")
    else:
        print(f"Status: WITHIN LIMIT")

    message_counter = message_counter + 1
    print()
```
