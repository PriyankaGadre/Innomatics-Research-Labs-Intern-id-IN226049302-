2. What happens if stopwords are not removed?

    Text becomes noisy and less meaningful

    Increases dimensionality

    Reduces model efficiency and accuracy

    Example: "I am going to the market" → stopwords add little meaning

   Scenario 1: Sentiment Analysis

         "I am not happy"

4. Stemming vs Lemmatization

| Feature | Stemming                    | Lemmatization     |
| ------- | --------------------------- | ----------------- |
| Method  | Rule-based chopping         | Dictionary-based  |
| Output  | Not always meaningful       | Always meaningful |
| Example | "running" → "run" or "runn" | "running" → "run" |


```python
import re
from collections import Counter
import numpy as np
```

```python
# Task 2: Preprocessing Function

def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)
    text = re.sub(r"\d+", "", text)
    text = text.lower()
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    cleaned_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ["no", "not"]
    ]

    cleaned_sentence = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_sentence


# 🔥 CALL FUNCTION
text = "I absolutely looooved this product 😍😍"

tokens, cleaned_sentence = preprocess_text(text)

print(tokens)
print(cleaned_sentence)
```

```python

# Task 3: Stress Testing

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

print("\n--- Stress Testing Output ---\n")
for sentence in test_sentences:
    tokens, clean_sentence = preprocess_text(sentence)

    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", clean_sentence)
    print("-"*50)

    results.append((tokens, clean_sentence))
```

```python

# Task 4: Token Analytics

print("\n--- Token Analytics ---\n")

analytics = []

for i, (tokens, _) in enumerate(results):
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(word) for word in tokens]) if tokens else 0

    print(f"Sentence {i+1}")
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length, 2))
    print("-"*40)

    analytics.append((total_tokens, unique_tokens, avg_length))
```

```python

# Task 5: Frequency Analysis

print("\n--- Frequency Analysis ---\n")

all_tokens = []

for tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

```python

# Task 6: Full Pipeline

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Example pipeline run
pipeline_output = full_pipeline(test_sentences)

print("\n--- Full Pipeline Output ---\n")
print(pipeline_output)
```

```python

# Task 7: Error Handling

print("\n--- Edge Case Testing ---\n")

edge_cases = [
    "",
    "😍😍😍😍",
    "1234567890"
]

for case in edge_cases:
    tokens, clean_sentence = preprocess_text(case)

    print("Input:", case)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", clean_sentence)
    print("-"*40)
```
